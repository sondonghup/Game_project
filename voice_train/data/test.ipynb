```python
import os
import torch
from scipy.io.wavfile import read
import librosa
import matplotlib.pyplot as plt
import numpy as np
```

```python
files = os.listdir('./')
for file in files:
    if file.endswith('.wav'):
        sampling_rate, data = read(file)
        print(sampling_rate)
        len = data.shape[0] / sampling_rate
        print('channel : ', data.shape[1])
        print(len)
        time = np.linspace(0, len, data.shape[0])
        plt.plot(time, data[:, 0])
        plt.plot(time, data[:, 1])
        plt.title(file)
        plt.show()
```

```python
files = os.listdir('./')
for file in files:
    if file.endswith('.wav'):
        data, sampling_rate = librosa.core.load(file)
        print(data.shape[0] / sampling_rate)
        len = data.shape[0] / sampling_rate
        # print('channel : ', data.shape[1])
        time = np.linspace(0, len, data.shape[0])
        print(time)
        plt.plot(time, data)
        # plt.plot(time, data[:, 1])
        plt.title(file)
        plt.show()
```
